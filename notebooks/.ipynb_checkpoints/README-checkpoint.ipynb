## 2 Evaluation of wavelets

```julia
plot(layout=(6,6),legend=false,ticks=nothing, border=nothing)
i=1;j=0;k=0;d=6
for p in 1:6
  qs = 2:2:6
  isodd(p) && (qs = 1:2:6 )
  for q in qs
    f, x = evaluate_in_dyadic_points(Primal, DWT.wavelet, DWT.CDFWavelet{p,q,Float64}(), j, k, d; points=true)
    plot!(x, -f, subplot=i)
    f, x = evaluate_in_dyadic_points(Primal, scaling, DWT.CDFWavelet{p,q,Float64}(), j, k, d; points=true)
    plot!(x, f, subplot=i)
    i += 1
  end
  for q in qs
    f, x = evaluate_in_dyadic_points(Dual, scaling, DWT.CDFWavelet{p,q,Float64}(), j, k, d; points=true)
    plot!(x, f, subplot=i)
    i += 1
  end
end
plot!()
```
